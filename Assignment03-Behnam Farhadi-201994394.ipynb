```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('olympic_athletes.csv')
```

```python
df.head(5)
```

```python
filter_df=df[(df['Season'].str.contains('Summer'))]
filter_df
              
```

```python
filter_df=filter_df[(filter_df['Event']=='Basketball Men\'s Basketball') & (filter_df['Medal']=='Gold')]
filter_df
```

```python
filter_df.describe()
```

```python
filter_df=filter_df[['Year','Height','Weight']].groupby('Year').mean().reset_index()
```

```python
# fig, ax1=plt.subplots()
# ax1.plot(filter_df['Year'], filter_df['Height'], color='black')

fig, ax = plt.subplots()

par = ax.twinx()

ax.set_xlabel("Year")
ax.set_ylabel("Height (cm)")
par.set_ylabel("Weight (kg)")
plt.title("Average Olympic Men's Gold Basketball Teams")
p1,=ax.plot(filter_df['Year'],filter_df['Height'] , label="Height", color='#000000', linewidth='1')
p2,=par.plot(filter_df['Year'],filter_df['Weight'], label="Weight", linestyle='--', color='r')
plt.legend([p1, p2],["Height", "Weight"])

ax.legend(loc='upper left')
plt.tight_layout()

plt.show()
plt.savefig('year-weight-height.jpg', dpi=600)
```
