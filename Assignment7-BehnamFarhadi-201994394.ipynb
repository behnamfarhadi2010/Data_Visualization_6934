### Question 1: (15 pts)
### The figure below shows the height distributions for the gold-medal winning athletes at the 2016 Olympics for several categories of sport. The points are the heights of individual athletes for the given sport.

```python
df = pd.read_csv('olympic_athletes.csv')
```

```python
df = df[(df['Year']== 2016) & (df['Medal']== 'Gold')]
df.drop_duplicates(subset='ID', keep='first')
```

```python
sss=df[(df['Sport']== 'Swimming') | (df['Sport']== 'Wrestling')|(df['Sport']== 'Fencing')| (df['Sport']== 'Athletics') | (df['Sport']== 'Shooting') | (df['Sport']== 'Cycling') | (df['Sport']== 'Rowing') | (df['Sport']== 'Gymnastics')]
sss.drop_duplicates(subset='ID', keep='first')
sss
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.set_title('Gold-medal Winning Heights at the 2016 Olympics')

ax = sns.violinplot(x="Sport", y="Height", data=sss, inner="quartile",scale="width")
ax = sns.swarmplot(x="Sport", y="Height", data=sss,
                   color="black")
ax.set_yticks([140,150,160,170,180,190,200,210,220])
```
